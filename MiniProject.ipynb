# Apple Store mini project
- Calculate the average rating of Gaming and Non-Gaming apps 
- Draw Pie plot based on labels as  below Price category and values as count of each category apps  
- labels:

    - Price =0  - Free apps
    - Price <20 - Affordable apps
    - Price<50 - Expensive apps
    - Price >50 - Very expensive

```python
import pandas as pd
filepath = 'AppleStore.csv'
def readCsvData(filepath):
    return pd.read_csv(filepath)
df = readCsvData(filepath)
df
```

```python
df.isnull().sum()
```

```python
a=pd.DataFrame(df['prime_genre'].value_counts())
a
```

## 1.Average user ratings for gaming and non gaming apps:

```python
#
x11=df['user_rating'][df.prime_genre=='Games'].mean()
print('Average Rating of gaming apps=',round(x11,2)) 
y11=df['user_rating'][df.prime_genre!='Games'].mean()
print('Average Rating of Non-Gaming apps',round(y11,2))
```

## 2.Pie chart based on Apps and their price
    - Free Apps
    - Affordable Apps
    - Expensive Apps
    - Very Expensive Apps

```python
li=[]
free=len(df[df['price']==0])
affordable=len(df[(df['price']!=0) & (df['price']<20)])
expensive=len(df[(df['price']>=20) & (df['price']<50)])
veryexpensive=len(df[df['price']>50])
li=[free,affordable,expensive,veryexpensive]
li
```

```python
import matplotlib.pyplot as plt
lang=['Free apps','Affordable apps','Expensive apps','Very expensive']
op=li
plt.pie(op,labels=lang,shadow=True,startangle=90,autopct='%1.1f%%')
plt.legend(loc='lower left')
plt.show()
```
